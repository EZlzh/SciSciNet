```python
import pandas as pd
date_str = '2021-12-06'
```

```python
%%time
df_MAG_PaperRef = pd.read_csv("./Data-Tables-MAG_" + date_str + "/df_MAG_PaperRef_whole.csv", sep='\t')
df_MAG_PaperRef
```

```python
%%time
df_MAG_Sub2Primary = pd.read_csv("./Data-Tables-MAG_" + date_str + "/df_MAG_Sub2Primary.csv", sep='\t')
df_MAG_Sub2Primary
```

```python
%%time
df_MAG_PaperRef_Sub = df_MAG_PaperRef.loc[(df_MAG_PaperRef['0'].isin(df_MAG_Sub2Primary['0'])) | 
                                          (df_MAG_PaperRef['1'].isin(df_MAG_Sub2Primary['0'])), :]
df_MAG_PaperRef_Sub
```

```python
%%time
import json
with open("./Data-Tables-MAG_" + date_str +"/MAG_Sub2Primary_Dict.json", 'r') as fp:
    MAG_Sub2Primary_Dict = json.load(fp)

MAG_Sub2Primary_Dict = {int(k):int(v) for k,v in MAG_Sub2Primary_Dict.items()}
print(len(MAG_Sub2Primary_Dict))
MAG_Sub2Primary_Dict[2257060365]
# Read_MAG_Paper_Raw.ipynb
```

```python
df_MAG_PaperRef_Sub_list = df_MAG_PaperRef_Sub[['0','1']].values
print(len(df_MAG_PaperRef_Sub_list))
df_MAG_PaperRef_Sub_list[:5]
```

```python
%%time
from tqdm import tqdm
for i,each in tqdm(enumerate(df_MAG_PaperRef_Sub_list)):
    if each[0] in MAG_Sub2Primary_Dict:
        df_MAG_PaperRef_Sub_list[i][0] = MAG_Sub2Primary_Dict[each[0]]
    if each[1] in MAG_Sub2Primary_Dict:
        df_MAG_PaperRef_Sub_list[i][1] = MAG_Sub2Primary_Dict[each[1]]
```

```python
print(len(df_MAG_PaperRef_Sub_list))
df_MAG_PaperRef_Sub_list[:5]
```

```python
%%time
df_MAG_PaperRef_Sub_New = pd.DataFrame(df_MAG_PaperRef_Sub_list)
df_MAG_PaperRef_Sub_New
```

```python
df_MAG_PaperRef_Sub_New[0]
```

```python
df_MAG_PaperRef_Sub_New.rename(columns={0: "0", 1: "1"},inplace=True)
df_MAG_PaperRef_Sub_New
```

```python
df_MAG_PaperRef_Sub_New['0']
```

```python
%%time
df_MAG_PaperRef_Pri = df_MAG_PaperRef.loc[(~df_MAG_PaperRef['0'].isin(df_MAG_Sub2Primary['0'])) & 
                                          (~df_MAG_PaperRef['1'].isin(df_MAG_Sub2Primary['0'])), :]
df_MAG_PaperRef_Pri
```

```python
%%time
df_MAG_PaperRef_RawSel = pd.concat([df_MAG_PaperRef_Pri, df_MAG_PaperRef_Sub_New], ignore_index=True)
df_MAG_PaperRef_RawSel
```

```python
%%time
# Drop duplicates
df_MAG_PaperRef_Sel = df_MAG_PaperRef_RawSel.drop_duplicates()
df_MAG_PaperRef_Sel
```
