```python
import pandas as pd
date_str = '2021-12-06'
```

```python
%%time
import csv
import pandas as pd

df_MAG_F = pd.read_csv('../../MAG_'+date_str+'/advanced/FieldsOfStudy.txt', sep='\t', header=None, 
                     quoting=csv.QUOTE_NONE)

df_MAG_F
```

```python
df_MAG_F_Sel = df_MAG_F.loc[df_MAG_F[5]<2,:]
df_MAG_F_Sel
```

```python
df_MAG_F_Lv0 = df_MAG_F.loc[df_MAG_F[5] == 0,:]
df_MAG_F_Lv0
```

```python
df_MAG_F_Sel = df_MAG_F_Sel.loc[:,[0,3,4,5,6,7,8]]
df_MAG_F_Sel
```

```python
def assign_name(num):
    if num == 0:
        return 'Top'
    if num == 1:
        return 'Sub'
df_MAG_F_Sel['Field_Type'] = df_MAG_F_Sel[5].apply(assign_name)
df_MAG_F_Sel
```

```python
df_MAG_F_Sel[df_MAG_F_Sel['Field_Type'] == 'Top']
```

```python
SciNet_Fields = df_MAG_F_Sel.loc[:,[0,3,'Field_Type']]
SciNet_Fields
```
