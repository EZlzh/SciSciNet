```python
import pandas as pd
date_str = '2021-12-06'
```

```python
%%time
df_MAG_PaperRef_Sel = pd.read_csv("./Data-Tables-MAG_" + date_str + "/df_MAG_PaperRef_Sel.csv", sep='\t')
df_MAG_PaperRef_Sel
# Link_MAG_Paper_Ref_Paper.ipynb
```

```python
%%time
df_MAG_Sel = pd.read_csv("./Data-Tables-MAG_" + date_str + "/df_MAG_Sel.csv", sep='\t')
df_MAG_Sel
# Sel_MAG_Paper.ipynb
```

```python
%%time
len(df_MAG_PaperRef_Sel['0'].unique())
```

```python
%%time
len(df_MAG_PaperRef_Sel['1'].unique())
```

```python
%%time
df_MAG_PaperRef_Sel_Y = df_MAG_PaperRef_Sel.merge(df_MAG_Sel[['0','7']].rename(columns={'7':'Year'}), on='0')
df_MAG_PaperRef_Sel_Y
```

```python
%%time
df_MAG_PaperRef_Sel_Y = df_MAG_PaperRef_Sel_Y.merge(
                        df_MAG_Sel[['0','7']].rename(columns={'7':'RefYear','0':'1'}), on='1')
df_MAG_PaperRef_Sel_Y
```

```python
%%time
df_MAG_PaperRef_Sel_Y['YearDiff'] = df_MAG_PaperRef_Sel_Y['Year'] - df_MAG_PaperRef_Sel_Y['RefYear']
df_MAG_PaperRef_Sel_Y
```

## Calculate C_all

```python
%%time
df_MAG_PaperRef_Sel_C = df_MAG_PaperRef_Sel_Y.groupby(['1'], 
                                                      as_index=False, sort=False)['0'].agg(['count']).reset_index()
df_MAG_PaperRef_Sel_C
```

```python
df_MAG_PaperRef_Sel_C.columns = ['0','C_all']
df_MAG_PaperRef_Sel_C
```

## Calculate C3, C5, C10

```python
%%time
df_MAG_PaperRef_Sel_C3 = df_MAG_PaperRef_Sel_Y.loc[(df_MAG_PaperRef_Sel_Y['YearDiff']>=0) & 
                                                   (df_MAG_PaperRef_Sel_Y['YearDiff']<=3),:].groupby(
    ['1'], as_index=False, sort=False)['0'].agg(['count']).reset_index()
df_MAG_PaperRef_Sel_C3
```

```python
df_MAG_PaperRef_Sel_C3.columns = ['0','C3']
df_MAG_PaperRef_Sel_C3
```

```python
%%time
df_MAG_PaperRef_Sel_C5 = df_MAG_PaperRef_Sel_Y.loc[(df_MAG_PaperRef_Sel_Y['YearDiff']>=0) & 
                                                   (df_MAG_PaperRef_Sel_Y['YearDiff']<=5),:].groupby(
    ['1'], as_index=False, sort=False)['0'].agg(['count']).reset_index()
df_MAG_PaperRef_Sel_C5.columns = ['0','C5']
df_MAG_PaperRef_Sel_C5
```

```python
%%time
df_MAG_PaperRef_Sel_C10 = df_MAG_PaperRef_Sel_Y.loc[(df_MAG_PaperRef_Sel_Y['YearDiff']>=0) & 
                                                   (df_MAG_PaperRef_Sel_Y['YearDiff']<=10),:].groupby(
    ['1'], as_index=False, sort=False)['0'].agg(['count']).reset_index()
df_MAG_PaperRef_Sel_C10.columns = ['0','C10']
df_MAG_PaperRef_Sel_C10
```

## Calculate R_all as Reference

```python
%%time
df_MAG_PaperRef_Sel_R = df_MAG_PaperRef_Sel_Y.groupby(['0'], 
                                                      as_index=False, sort=False)['1'].agg(['count']).reset_index()
df_MAG_PaperRef_Sel_R.columns = ['0','R']
df_MAG_PaperRef_Sel_R
```

## Merge these records to MAG main table

```python
%%time
df_MAG_Sel_New = df_MAG_Sel.merge(df_MAG_PaperRef_Sel_C, on='0', how='left')
df_MAG_Sel_New
```

```python
%%time
df_MAG_Sel_New = df_MAG_Sel_New.merge(df_MAG_PaperRef_Sel_C10, on='0', how='left')
df_MAG_Sel_New = df_MAG_Sel_New.merge(df_MAG_PaperRef_Sel_R, on='0', how='left')
df_MAG_Sel_New
```
