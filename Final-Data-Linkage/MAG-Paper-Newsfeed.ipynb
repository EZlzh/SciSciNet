```python
import pandas as pd
date_str = '2021-12-06'
```

```python
%%time
df_MAG_Sel_wRC = pd.read_csv("../Final-Data/Data-Tables-MAG_" + date_str + "/df_MAG_Sel_wRC.csv", sep='\t')
df_MAG_Sel_wRC
# Sel_MAG_Ref_Cite.ipynb
```

```python
df_MAG_Newsfeed = pd.read_csv('../../DataLake/Newsfeed/DOI_Newsfeed_Raw.csv', sep='\t')
df_MAG_Newsfeed
# Read papers with Newsfeed.
# Crawled from https://www.eventdata.crossref.org/guide/sources/newsfeed/
# Wrong name: should be called df_DOI_Newsfeed.
# But it does not matter.
```

```python
%%time
df_MAG_Newsfeed.loc[df_MAG_Newsfeed['obj_id'].astype(str).str.startswith('https://doi.org/'), :]
```

```python
prefix = 'https://doi.org/'
def remove_prefix(text):
    if text.startswith(prefix):
        return text[len(prefix):]
    return text  # or whatever
remove_prefix('https://doi.org/10.1038/nature21680')
```

```python
%%time
df_MAG_Newsfeed['DOI'] = df_MAG_Newsfeed['obj_id'].apply(remove_prefix)
df_MAG_Newsfeed
```

```python
df_MAG_Newsfeed['DOI'] = df_MAG_Newsfeed['DOI'].str.lower()
df_MAG_Newsfeed
```

```python
%%time
df_MAG_Sel_DOI = df_MAG_Sel_wRC.loc[pd.notnull(df_MAG_Sel_wRC['DOI']),['MAGID','DOI']]
df_MAG_Sel_DOI
```

```python
df_MAG_Sel_DOI['DOI'] = df_MAG_Sel_DOI['DOI'].str.lower()
df_MAG_Sel_DOI
```

```python
%%time
df_MAG_Newsfeed.loc[df_MAG_Newsfeed['DOI'].isin(df_MAG_Sel_DOI['DOI']),:]
```

```python
len(df_MAG_Sel_DOI['MAGID'].unique())
```

```python
len(df_MAG_Sel_DOI['DOI'].unique())
```

```python
%%time
df_MAG_Sel_DOI_cnt = df_MAG_Sel_DOI.groupby(['DOI'],as_index=False)['MAGID'].agg(
    ['count']).reset_index().sort_values(['count'], ascending=False)
df_MAG_Sel_DOI_cnt
```

```python
df_MAG_Sel_DOI_cnt_LargerThan2 = df_MAG_Sel_DOI_cnt.loc[df_MAG_Sel_DOI_cnt['count']>1,:]
df_MAG_Sel_DOI_cnt_LargerThan2
```

```python
df_MAG_Newsfeed.loc[df_MAG_Newsfeed['DOI'].isin(df_MAG_Sel_DOI_cnt_LargerThan2['DOI']),:]
```

```python
df_MAG_Sel_DOI_cnt_LargerThan5 = df_MAG_Sel_DOI_cnt.loc[df_MAG_Sel_DOI_cnt['count']>5,:]
df_MAG_Sel_DOI_cnt_LargerThan5
```

```python
df_MAG_Newsfeed.loc[df_MAG_Newsfeed['DOI'].isin(df_MAG_Sel_DOI_cnt_LargerThan5['DOI']),:]
```

```python
%%time
df_MAG_Newsfeed_Sel = df_MAG_Newsfeed.merge(df_MAG_Sel_DOI, how='inner', on='DOI')
df_MAG_Newsfeed_Sel
# Merge the Primary MAG papers with Newsfeed via DOI exact matching.
```

# Read MAG to Newsfeed Count

```python
df_MAG_Newsfeed_Final_New = pd.read_csv('./Data-Linkage/Link_MAG_Newsfeed.csv', sep='\t')
df_MAG_Newsfeed_Final_New
```

```python
%%time
df_MAG_Newsfeed_cnt = df_MAG_Newsfeed_Final_New.groupby(['MAGID'],as_index=False)['subj_id'].agg(['count']).reset_index().sort_values(['count'], ascending=False)
df_MAG_Newsfeed_cnt
# Counting how many news cite a paper
```

# Read MAGID Newsfeed Count

```python
df_MAG_Newsfeed_cnt = pd.read_csv('./Data-Linkage/MAGID_Newsfeedcount.csv', sep='\t')
df_MAG_Newsfeed_cnt
```

```python
%%time
df_MAG_Newsfeed_cnt1 = df_MAG_Newsfeed_cnt.groupby(['Newsfeed_count'])['MAGID'].agg(
    ['count']).reset_index().sort_values(['count'], ascending=False)
df_MAG_Newsfeed_cnt1
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
colors = [i for i in cm.get_cmap('tab20').colors]
fig, ax1 = plt.subplots(figsize=(12, 8))
x = np.arange(20) + 1
ax1.bar(x, 
        df_MAG_Newsfeed_cnt1.head(20)['count'].values,
        color=colors[4])
ax1.set_xlabel('News citation counts', fontsize=16)
ax1.set_ylabel('MAG Paper counts', fontsize=16)
ax1.set_yscale('log')
ax1.tick_params(axis='y', labelsize=12)
ax1.tick_params(axis='x', labelsize=12)
# ax1.set_xscale('log')
```
