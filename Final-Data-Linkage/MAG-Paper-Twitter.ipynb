```python
prefix = 'https://doi.org/'
def remove_prefix(text):
    if text.startswith(prefix):
        return text[len(prefix):]
    return text  # or whatever
remove_prefix('https://doi.org/10.1098/rspb.2016.0792')
```

```python
%%time
df_DOI_Twitter['DOI'] = df_DOI_Twitter['obj_id'].apply(remove_prefix)
df_DOI_Twitter
```

```python
%%time
df_DOI_Twitter['DOI'] = df_DOI_Twitter['DOI'].str.lower()
df_DOI_Twitter
```

# Read MAG to Twitter Count

```python
%%time
df_MAG_Twitter_New = pd.read_csv('./Data-Linkage/Link_MAG_Twitter.csv', sep='\t')
df_MAG_Twitter_New
```

```python
len(df_MAG_Twitter_New['MAGID'].unique())
```

```python
%%time
df_MAG_Twitter_cnt = df_MAG_Twitter_New.groupby(['MAGID'],as_index=False)['subj_id'].agg(['count']).reset_index().sort_values(['count'], ascending=False)
df_MAG_Twitter_cnt
# Counting how many news cite a paper
```

```python
df_MAG_Twitter_cnt.rename(columns={'count':'Twitter_count'}, inplace=True)
df_MAG_Twitter_cnt
```

# Read MAGID Twitter Count

```python
df_MAG_Twitter_cnt = pd.read_csv('./Data-Linkage/MAGID_Twittercount.csv', sep='\t')
df_MAG_Twitter_cnt
```

```python
%%time
df_MAG_Twitter_cnt1 = df_MAG_Twitter_cnt.groupby(['Twitter_count'])['MAGID'].agg(
    ['count']).reset_index().sort_values(['count'], ascending=False)
df_MAG_Twitter_cnt1
```

```python
df_MAG_Twitter_cnt1.head(20)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
colors = [i for i in cm.get_cmap('tab20').colors]
fig, ax1 = plt.subplots(figsize=(12, 8))
x = np.arange(20) + 1
ax1.bar(x, 
        df_MAG_Twitter_cnt1.head(20)['count'].values,
        color=colors[6])
ax1.set_xlabel('Twitter mention counts', fontsize=16)
ax1.set_ylabel('MAG Paper counts', fontsize=16)
ax1.set_yscale('log')
ax1.tick_params(axis='y', labelsize=12)
ax1.tick_params(axis='x', labelsize=12)
# ax1.set_xscale('log')
```
