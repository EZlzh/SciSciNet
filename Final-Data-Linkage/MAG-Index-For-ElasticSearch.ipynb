```python
from collections import *
MAG_PaperID_Authors_Dict = defaultdict(list)
```

```python
from tqdm import tqdm
for each_r in tqdm(MAG_PaperID_Author):
    MAG_PaperID_Authors_Dict[each_r[0]].append(each_r[1])
```

```python
MAG_PaperID_Authors_Dict[3126147277]
```

```python
temp = MAG_PaperID_Authors_Dict[3126147277].copy()
temp2 = ', '.join(temp)
temp2
```

```python
for k,v in tqdm(MAG_PaperID_Authors_Dict.items()):
    MAG_PaperID_Authors_Dict[k] = ', '.join(v)
```

```python
MAG_PaperID_Authors_Dict[3126147277]
```

```python
from tqdm import tqdm
MAG_PaperID_Authors_List = []
for k,v in tqdm(MAG_PaperID_Authors_Dict.items()):
    MAG_PaperID_Authors_List.append([k,v])
#     break
```
