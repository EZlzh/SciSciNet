```python
import pandas as pd
date_str = '2021-12-06'
```

```python
%%time
import csv
import pandas as pd

df_MAG_Ext = pd.read_csv('../../MAG_'+date_str+'/mag/PaperExtendedAttributes.txt', sep='\t', header=None, 
                     quoting=csv.QUOTE_NONE)

df_MAG_Ext
```

```python
%%time
df_MAG_PubMed = df_MAG_Ext.loc[df_MAG_Ext[1] == 2, :]
df_MAG_PubMed
```

```python
%%time
df_MAG_Sel_wRC = pd.read_csv("../Final-Data/Data-Tables-MAG_" + date_str + "/df_MAG_Sel_wRC.csv", sep='\t')
df_MAG_Sel_wRC
# Sel_MAG_Ref_Cite.ipynb
```

```python
df_MAG_Sub2Primary = pd.read_csv("../Final-Data/Data-Tables-MAG_" + date_str + "/df_MAG_Sub2Primary.csv", sep='\t')
df_MAG_Sub2Primary
# Read_MAG_Paper_Raw.ipynb
```

```python
%%time
import json
with open("../Final-Data/Data-Tables-MAG_" + date_str +"/MAG_Sub2Primary_Dict.json", 'r') as fp:
    MAG_Sub2Primary_Dict = json.load(fp)

MAG_Sub2Primary_Dict = {int(k):int(v) for k,v in MAG_Sub2Primary_Dict.items()}
print(len(MAG_Sub2Primary_Dict))
MAG_Sub2Primary_Dict[2257060365]
# Read_MAG_Paper_Raw.ipynb
```

```python
# Deal with PubMed ID (PMID).
```

```python
%%time
df_MAG_PubMed_Sel = df_MAG_PubMed.loc[(df_MAG_PubMed[0].isin(df_MAG_Sel_wRC['MAGID'])) | 
                                      (df_MAG_PubMed[0].isin(df_MAG_Sub2Primary['0'])), :]
df_MAG_PubMed_Sel
```

```python
print("Proportion:,", len(df_MAG_PubMed_Sel) / len(df_MAG_PubMed))
```

```python
%%time
df_MAG_PubMed_Pri = df_MAG_PubMed_Sel.loc[(~df_MAG_PubMed_Sel[0].isin(df_MAG_Sub2Primary['0'])), :]
df_MAG_PubMed_Pri
```

```python
%%time
df_MAG_PubMed_Raw = df_MAG_PubMed_Sel.loc[(df_MAG_PubMed_Sel[0].isin(df_MAG_Sub2Primary['0'])), :]
df_MAG_PubMed_Raw
```

```python
df_MAG_PubMed_Raw_list = df_MAG_PubMed_Raw[[0,1,2]].values
print(len(df_MAG_PubMed_Raw_list))
df_MAG_PubMed_Raw_list[:5]
```

```python
%%time
from tqdm import tqdm
for i,each in tqdm(enumerate(df_MAG_PubMed_Raw_list)):
    if each[0] in MAG_Sub2Primary_Dict:
        df_MAG_PubMed_Raw_list[i][0] = MAG_Sub2Primary_Dict[each[0]]
```

```python
print(len(df_MAG_PubMed_Raw_list))
df_MAG_PubMed_Raw_list[:5]
```

```python
%%time
df_MAG_PubMed_Sub_New = pd.DataFrame(df_MAG_PubMed_Raw_list)
df_MAG_PubMed_Sub_New
```

```python
%%time
df_MAG_PaperRef_RawSel = pd.concat([df_MAG_PubMed_Pri, df_MAG_PubMed_Sub_New], ignore_index=True)
df_MAG_PaperRef_RawSel
```

```python
%%time
# Drop duplicates
df_MAG_PaperRef_Sel = df_MAG_PaperRef_RawSel.drop_duplicates()
df_MAG_PaperRef_Sel
```

```python
df_MAG_PaperRef_Sel.drop([1], axis=1, inplace=True)
df_MAG_PaperRef_Sel
```
