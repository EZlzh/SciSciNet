```python
import pandas as pd
date_str = '2021-12-06'
```

```python
%%time
df_MAG_Sel_wRC = pd.read_csv("../Final-Data/Data-Tables-MAG_" + date_str + "/df_MAG_Sel_wRC.csv", sep='\t')
df_MAG_Sel_wRC
# Sel_MAG_Ref_Cite.ipynb
```

```python
%%time
import numpy as np
SciNet_Papers = df_MAG_Sel_wRC.loc[:,['MAGID','DOI','DocType','Year','Date','JID','ConfSID',
                                      'C_all','C10','R','C5']]
SciNet_Papers.rename(columns={'MAGID':'PaperID', 'JID':'JournalID', 'ConfSID':'ConferenceSeriesID',
                             'C_all':'Citation_Count', 'R':'Reference_Count'}, inplace=True)
SciNet_Papers.loc[SciNet_Papers['Year'] >= 2012, ['C10']] = np.nan
SciNet_Papers.loc[SciNet_Papers['Year'] >= 2017, ['C5']] = np.nan
SciNet_Papers
```

```python
SciNet_Papers
```

```python
%%time
SciNet_Team_Institution = pd.read_csv(
    '../Final-Data-Metrics/Data-Count/SciNet_Team_Institution.tsv', sep='\t')
SciNet_Team_Institution
```

```python
%%time
SciNet_Papers = SciNet_Papers.merge(SciNet_Team_Institution, how='left', on='PaperID')
SciNet_Papers
```

```python
%%time
df_Disruption_Sel = pd.read_csv('../Final-Data-Metrics/Data-Count/PaperID_Disruption.tsv', sep='\t')
df_Disruption_Sel
```

```python
%%time
df_Atyp_Comb = pd.read_csv('../Final-Data-Metrics/Data-Count/SciNet-Atyp-Comb-Final.tsv', sep='\t')
df_Atyp_Comb
```

```python
%%time
df_SB = pd.read_csv('../Final-Data-Metrics/Data-Count/SciNet-SleepingBeauty.tsv', sep='\t')
df_SB
```

```python
%%time
SciNet_Papers = SciNet_Papers.merge(df_Disruption_Sel, how='left', on='PaperID')
SciNet_Papers = SciNet_Papers.merge(df_Atyp_Comb, how='left', on='PaperID')
SciNet_Papers = SciNet_Papers.merge(df_SB, how='left', on='PaperID')
SciNet_Papers
```

```python
%%time
SciNet_Link_Patent = pd.read_csv('../Final-SciNet-Tables/SciNet_Link_Paper_Patents.tsv', sep='\t')
SciNet_Link_Patent = SciNet_Link_Patent.drop_duplicates()
SciNet_Link_Patent_cnt = SciNet_Link_Patent.groupby(
    ['PaperID'],as_index=False)['PatentID'].agg(['count']).reset_index().sort_values(['count'], ascending=False)
SciNet_Link_Patent_cnt.rename(columns={'count':'Patent_Count'}, inplace=True)
SciNet_Link_Patent_cnt
```

```python
%%time
SciNet_Link_Newsfeed = pd.read_csv('../Final-SciNet-Tables/SciNet_Link_Paper_Newsfeed.tsv', sep='\t')
SciNet_Link_Newsfeed = SciNet_Link_Newsfeed.drop_duplicates()
SciNet_Link_Newsfeed_cnt = SciNet_Link_Newsfeed.groupby(
    ['PaperID'],as_index=False)['NewsfeedID'].agg(['count']).reset_index().sort_values(['count'], ascending=False)
SciNet_Link_Newsfeed_cnt.rename(columns={'count':'Newsfeed_Count'}, inplace=True)
SciNet_Link_Newsfeed_cnt
```

```python
%%time
SciNet_Link_Twitter = pd.read_csv('../Final-SciNet-Tables/SciNet_Link_Paper_Twitter.tsv', sep='\t')
SciNet_Link_Twitter = SciNet_Link_Twitter.drop_duplicates()
SciNet_Link_Twitter_cnt = SciNet_Link_Twitter.groupby(
    ['PaperID'],as_index=False)['TweetID'].agg(['count']).reset_index().sort_values(['count'], ascending=False)
SciNet_Link_Twitter_cnt.rename(columns={'count':'Twitter_Count'}, inplace=True)
SciNet_Link_Twitter_cnt
```

```python
%%time
SciNet_Link_ClinicalTrials = pd.read_csv('../Final-SciNet-Tables/SciNet_Link_Paper_ClinicalTrials.tsv', sep='\t')
SciNet_Link_ClinicalTrials = SciNet_Link_ClinicalTrials.drop_duplicates()
SciNet_Link_ClinicalTrials_cnt = SciNet_Link_ClinicalTrials.groupby(
    ['PaperID'],as_index=False)['NCT_Number'].agg(['count']).reset_index().sort_values(['count'], ascending=False)
SciNet_Link_ClinicalTrials_cnt.rename(columns={'count':'NCT_Count'}, inplace=True)
SciNet_Link_ClinicalTrials_cnt
```

```python
%%time
SciNet_Link_NIH = pd.read_csv('../Final-SciNet-Tables/SciNet_Link_Paper_NIH.tsv', sep='\t')
SciNet_Link_NIH = SciNet_Link_NIH.drop_duplicates()
SciNet_Link_NIH_cnt = SciNet_Link_NIH.groupby(
    ['PaperID'],as_index=False)['NIH_Project_Number'].agg(['count']).reset_index().sort_values(['count'], ascending=False)
SciNet_Link_NIH_cnt.rename(columns={'count':'NIH_Count'}, inplace=True)
SciNet_Link_NIH_cnt
```

```python
%%time
SciNet_Link_NSF = pd.read_csv('../Final-SciNet-Tables/SciNet_Link_Paper_NSF.tsv', sep='\t')
SciNet_Link_NSF = SciNet_Link_NSF.drop_duplicates(subset=['NSF_Award_Number','PaperID'],keep='first')
SciNet_Link_NSF_cnt = SciNet_Link_NSF.groupby(
    ['PaperID'],as_index=False)['NSF_Award_Number'].agg(['count']).reset_index().sort_values(['count'], ascending=False)
SciNet_Link_NSF_cnt.rename(columns={'count':'NSF_Count'}, inplace=True)
SciNet_Link_NSF_cnt
```

```python
%%time
SciNet_Papers = SciNet_Papers.merge(SciNet_Link_Patent_cnt, how='left', on='PaperID')
SciNet_Papers = SciNet_Papers.merge(SciNet_Link_Newsfeed_cnt, how='left', on='PaperID')
SciNet_Papers = SciNet_Papers.merge(SciNet_Link_Twitter_cnt, how='left', on='PaperID')
SciNet_Papers = SciNet_Papers.merge(SciNet_Link_ClinicalTrials_cnt, how='left', on='PaperID')
SciNet_Papers = SciNet_Papers.merge(SciNet_Link_NIH_cnt, how='left', on='PaperID')
SciNet_Papers = SciNet_Papers.merge(SciNet_Link_NSF_cnt, how='left', on='PaperID')
SciNet_Papers
```

```python
%%time
SciNet_Papers['Patent_Count'].fillna(0, inplace=True)
SciNet_Papers['Newsfeed_Count'].fillna(0, inplace=True)
SciNet_Papers['Twitter_Count'].fillna(0, inplace=True)
SciNet_Papers['NCT_Count'].fillna(0, inplace=True)
SciNet_Papers['NIH_Count'].fillna(0, inplace=True)
SciNet_Papers['NSF_Count'].fillna(0, inplace=True)
SciNet_Papers
```

```python
%%time
SciNet_Papers['Patent_Count'] = SciNet_Papers['Patent_Count'].astype(int)
SciNet_Papers['Newsfeed_Count'] = SciNet_Papers['Newsfeed_Count'].astype(int)
SciNet_Papers['Twitter_Count'] = SciNet_Papers['Twitter_Count'].astype(int)
SciNet_Papers['NCT_Count'] = SciNet_Papers['NCT_Count'].astype(int)
SciNet_Papers['NIH_Count'] = SciNet_Papers['NIH_Count'].astype(int)
SciNet_Papers['NSF_Count'] = SciNet_Papers['NSF_Count'].astype(int)
SciNet_Papers
```

```python
%%time
SciNet_Papers.rename(columns={'Twitter_Count':'Tweet_Count'},inplace=True)
SciNet_Papers
```
