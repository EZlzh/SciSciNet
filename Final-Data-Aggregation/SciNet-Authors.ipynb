```python
import pandas as pd
```

```python
%%time
SciNet_PaperAuthorAffiliations = pd.read_csv(
    '../Final-SciNet-Tables/SciNet_PaperAuthorAffiliations.tsv', sep='\t')
SciNet_PaperAuthorAffiliations
```

```python
date_str = "2021-12-06"
```

```python
%%time
import csv
import pandas as pd

df_MAG_Authors = pd.read_csv('../../MAG_'+date_str+'/mag/Authors.txt', sep='\t', header=None, quoting=csv.QUOTE_NONE, 
                      usecols=[0,2,3])
df_MAG_Authors
```

```python
%%time
df_MAG_Authors.rename(columns={0:'AuthorID',2:'Normalized_Name',3:'Author_Name'},inplace=True)
df_MAG_Authors
```

```python
%%time
df_MAG_Authors_Sel = df_MAG_Authors.loc[
    df_MAG_Authors['AuthorID'].isin(SciNet_PaperAuthorAffiliations['AuthorID']), :]
df_MAG_Authors_Sel
```

```python
%%time
len(df_MAG_Authors_Sel['AuthorID'].unique())
```

```python
df_MAG_Authors_Sel
```

```python
%%time
df_Author_Prod = pd.read_csv('../Final-Data-Metrics/Data-Count/SciNet-Prod.tsv', sep='\t')
df_Author_Prod
```

```python
%%time
df_Hindex = pd.read_csv('../Final-Data-Metrics/Data-Count/SciNet-H-Index.tsv', sep='\t')
df_Hindex
```

```python
%%time
df_AvgC10 = pd.read_csv('../Final-Data-Metrics/Data-Count/SciNet-Average-C10.tsv', sep='\t')
df_AvgC10
```

```python
%%time
df_MAG_Authors_Sel1 = df_MAG_Authors_Sel[['AuthorID','Author_Name']].merge(
    df_Hindex, how='left', on='AuthorID')
df_MAG_Authors_Sel1
```

```python
%%time
df_MAG_Authors_Sel2 = df_MAG_Authors_Sel1.merge(
    df_Author_Prod, how='left', on='AuthorID')
df_MAG_Authors_Sel2
```

```python
%%time
df_MAG_Authors_Sel3 = df_MAG_Authors_Sel2.merge(
    df_AvgC10, how='left', on='AuthorID'
)
df_MAG_Authors_Sel3
```

```python
%%time
len(df_MAG_Authors_Sel3[pd.notnull(df_MAG_Authors_Sel3['Average_C10'])])
```

```python
%%time
len(df_MAG_Authors_Sel3[pd.notnull(df_MAG_Authors_Sel3['Average_LogC10'])])
```

```python
df_MAG_Authors_Sel3.rename(columns={'H-Index':'H-index'},inplace=True)
df_MAG_Authors_Sel3
```

```python
df_MAG_Authors_Sel3
```

```python
df_MAG_Authors_Sel3.to_csv('../Final-SciNet-Tables/SciNet_Authors.tsv', sep='\t', index=False)
```
