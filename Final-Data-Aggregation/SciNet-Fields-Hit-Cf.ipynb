```python
import pandas as pd
date_str = '2021-12-06'
```

```python
%%time
df_MAG_PF_Lv0_Sel = pd.read_csv("../Final-Data/Data-Tables-MAG_" + date_str + "/df_MAG_PF_Lv0_Sel.csv", sep='\t')
df_MAG_PF_Lv0_Sel
# Read_MAG_Paper_Fields.ipynb
```

```python
%%time
df_MAG_PF_Lv1_Sel = pd.read_csv("../Final-Data/Data-Tables-MAG_" + date_str + "/df_MAG_PF_Lv1_Sel.csv", sep='\t')
df_MAG_PF_Lv1_Sel
# Read_MAG_Paper_Fields.ipynb
```

```python
df_MAG_PF_Lv0_Hit_Sel = pd.read_csv('../Final-Data-Metrics/Data-Metrics/MAG_PF_Lv0_Hit.csv', sep='\t')
df_MAG_PF_Lv0_Hit_Sel
```

```python
df_MAG_PF_Lv1_Hit_Sel = pd.read_csv('../Final-Data-Metrics/Data-Metrics/MAG_PF_Lv1_Hit.csv', sep='\t')
df_MAG_PF_Lv1_Hit_Sel
```

```python
df_MAG_Lv0_Cf_Sel = pd.read_csv('../Final-Data-Metrics/Data-Metrics/MAG_PF_Lv0_Cf.csv', sep='\t')
df_MAG_Lv0_Cf_Sel
```

```python
df_MAG_Lv1_Cf_Sel = pd.read_csv('../Final-Data-Metrics/Data-Metrics/MAG_PF_Lv1_Cf.csv', sep='\t')
df_MAG_Lv1_Cf_Sel
```

```python
df_MAG_PF_Lv0_Sel.columns = ['MAGID','FieldID','Score']
df_MAG_PF_Lv1_Sel.columns = ['MAGID','FieldID','Score']
```

```python
df_MAG_PF_Lv0_Sel
```

```python
df_MAG_PF_Lv0_Hit_Sel
```

```python
%%time
df_MAG_PF_Lv0_New = df_MAG_PF_Lv0_Sel[['MAGID','FieldID']].merge(df_MAG_PF_Lv0_Hit_Sel, 
                                                                 how='left', on=['MAGID','FieldID'])
df_MAG_PF_Lv0_New
```

```python
%%time
df_MAG_PF_Lv0_New = df_MAG_PF_Lv0_New.merge(df_MAG_Lv0_Cf_Sel,
                                            how='left', on=['MAGID','FieldID'])
df_MAG_PF_Lv0_New
```

```python
%%time
df_MAG_PF_Lv1_New = df_MAG_PF_Lv1_Sel[['MAGID','FieldID']].merge(df_MAG_PF_Lv1_Hit_Sel, 
                                                                 how='left', on=['MAGID','FieldID'])
df_MAG_PF_Lv1_New
```

```python
%%time
df_MAG_PF_Lv1_New = df_MAG_PF_Lv1_New.merge(df_MAG_Lv1_Cf_Sel,
                                            how='left', on=['MAGID','FieldID'])
df_MAG_PF_Lv1_New
```

```python
df_MAG_PF_Lv0_New
```

```python
df_MAG_PF_Lv0_New.columns = ['PaperID', 'FieldID', 'Hit_1pct', 'Hit_5pct', 'Hit_10pct', 'C_f']
df_MAG_PF_Lv0_New
```

```python
df_MAG_PF_Lv1_New
```

```python
df_MAG_PF_Lv1_New.columns = ['PaperID', 'FieldID', 'Hit_1pct', 'Hit_5pct', 'Hit_10pct', 'C_f']
df_MAG_PF_Lv1_New
```

```python
%%time
SciNet_PaperFields = pd.concat([df_MAG_PF_Lv0_New,df_MAG_PF_Lv1_New], ignore_index=True)
SciNet_PaperFields
```
