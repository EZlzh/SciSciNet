```python
import pandas as pd
date_str = '2021-12-06'
```

```python
%%time
import csv
import pandas as pd

df_MAG = pd.read_csv('../../MAG_'+date_str+'/mag/Papers.txt', sep='\t', header=None, quoting=csv.QUOTE_NONE, 
                      usecols=[0,2,3,5,6,7,8,10,11,12,14,15,16,17,18,19,21,22,24])
# https://docs.microsoft.com/en-us/academic-services/graph/reference-data-schema#papers
# Read MAG data.
# 0:MAG_PaperId; 2:DOI; 3:DocType; 5:OriginalPaperTitle; 6:BookTitle;
# 7:Year; 8:Date; 10:Publisher; 11:JournalId; 12:ConferenceSeriesId; 
# 14:Volume; 15:Issue; 16:FirstPage; 17:LastPage;
# 18:ReferenceCount; 19:CitationCount; 20:EstimatedCitation; 
# 21:OriginalVenue; 22:FamilyId; 24:DocSubTypes for retraction.

df_MAG
```

```python
%%time
df_MAG = df_MAG.loc[((pd.notnull(df_MAG[3])) | (pd.notnull(df_MAG[2]))), :]
# Delete papers that had neither DOI nor DocType.

df_MAG = df_MAG.loc[(df_MAG[3] != 'Patent'), :]
# Delete papers with DocType as 'Patent'.

df_MAG
```
