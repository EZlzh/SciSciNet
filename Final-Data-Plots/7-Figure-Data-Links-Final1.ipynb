```python
import pandas as pd
```

```python
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
Chunks = []
Chunks_Name = []
Chunks_CountName = []
Chunks_Total = []
```

```python
%%time
SciNet_Link_Twitter = pd.read_csv('../Final-SciNet-Tables/SciNet_Link_Twitter.tsv', sep='\t')
SciNet_Link_Twitter_cnt = SciNet_Link_Twitter.groupby(
    ['PaperID'],as_index=False)['TweetID'].agg(['count']).reset_index().sort_values(['count'], ascending=False)
SciNet_Link_Twitter_cnt.rename(columns={'count':'Twitter_Count'}, inplace=True)
SciNet_Link_Twitter_cnt
```

```python
%%time
SciNet_Link_Twitter_cnt1 = SciNet_Link_Twitter_cnt.groupby(['Twitter_Count'])['PaperID'].agg(
    ['count']).reset_index().sort_values(['count'], ascending=False)
SciNet_Link_Twitter_cnt1
```

```python
Chunks.append(SciNet_Link_Twitter_cnt1)
Chunks_Name.append('Twitter Mentions')
Chunks_CountName.append('Twitter_Count')
Chunks_Total.append(SciNet_Link_Twitter_cnt)
```

```python
Chunks.append(SciNet_Link_Newsfeed_cnt1)
Chunks_Name.append('Newsfeed Mentions')
Chunks_CountName.append('Newsfeed_Count')
Chunks_Total.append(SciNet_Link_Newsfeed_cnt)
```

```python
%%time
SciNet_Link_NSF = pd.read_csv('../Final-SciNet-Tables/SciNet_Link_NSF.tsv', sep='\t')
SciNet_Link_NSF_cnt = SciNet_Link_NSF.groupby(
    ['PaperID'],as_index=False)['NSF_Award_Number'].agg(['count']).reset_index().sort_values(['count'], ascending=False)
SciNet_Link_NSF_cnt.rename(columns={'count':'NSF_Count'}, inplace=True)
SciNet_Link_NSF_cnt
```

```python
%%time
SciNet_Link_NSF_cnt1 = SciNet_Link_NSF_cnt.groupby(['NSF_Count'])['PaperID'].agg(
    ['count']).reset_index().sort_values(['count'], ascending=False)
SciNet_Link_NSF_cnt1.head(5)
```

```python
Chunks.append(SciNet_Link_NSF_cnt1)
Chunks_Name.append('NSF Awards')
Chunks_CountName.append('NSF_Count')
Chunks_Total.append(SciNet_Link_NSF_cnt)
```

```python
%%time
%matplotlib inline
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
colors = [i for i in cm.get_cmap('tab20').colors]
fig, ax1 = plt.subplots(2,3, figsize=(16, 10))
ax1 = ax1.ravel()
i = 0
for i in range(6):
    cnt_name = Chunks_CountName[i]
#     print(cnt_name)
#     print(np.min(Chunks[i][cnt_name].values), np.max(Chunks[i][cnt_name].values))
    binnum = int(len(Chunks[i][cnt_name]) / 3)
    if binnum > 50:
        binnum = 30
#     print(binnum)
#     print(np.min(Chunks[i][cnt_name].values),np.max(Chunks[i][cnt_name].values))
    bins = np.logspace(np.log10(np.min(Chunks[i][cnt_name].values)), 
                   np.log10(np.max(Chunks[i][cnt_name].values)),
                   num=binnum)
#     print(bins)
#     print(bins[0],bins[-1])
    hist, _ = np.histogram(Chunks_Total[i][cnt_name].values, bins=bins)
    h_xaxis = 10**((np.log10(bins[:-1]) + np.log10(bins[1:])) / 2)
#     print(len(h_xaxis))
    ax1[i].plot(h_xaxis, hist, 'o', ms=8, mfc='none', color=colors[i*2+1], markeredgewidth=2)
    ax1[i].set_ylabel('# of Papers', fontsize=16)
    ax1[i].set_xlabel('# of ' + Chunks_Name[i], fontsize=16)
    ax1[i].set_yscale('log')
    ax1[i].set_xscale('log')
    ax1[i].tick_params(axis='y', labelsize=12)
    ax1[i].tick_params(axis='x', labelsize=12)
    ax1[i].spines['top'].set_visible(False)
    ax1[i].spines['right'].set_visible(False)
plt.subplots_adjust(wspace=0.2, hspace=0.4)
```

```python
%%time
%matplotlib inline
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
colors = [i for i in cm.get_cmap('tab20').colors]
fig, ax1 = plt.subplots(2,3, figsize=(16, 10))
ax1 = ax1.ravel()
i = 0
for i in range(6):
    cnt_name = Chunks_CountName[i]
#     print(cnt_name)
#     print(np.min(Chunks[i][cnt_name].values), np.max(Chunks[i][cnt_name].values))
    binnum = int(len(Chunks[i][cnt_name]) / 3)
    if binnum > 50:
        binnum = 30
#     print(binnum)
#     print(np.min(Chunks[i][cnt_name].values),np.max(Chunks[i][cnt_name].values))
    bins = np.logspace(np.log10(np.min(Chunks[i][cnt_name].values)), 
                   np.log10(np.max(Chunks[i][cnt_name].values)),
                   num=binnum)
#     print(bins)
#     print(bins[0],bins[-1])
    hist, _ = np.histogram(Chunks_Total[i][cnt_name].values, bins=bins)
    h_xaxis = 10**((np.log10(bins[:-1]) + np.log10(bins[1:])) / 2)
#     print(len(h_xaxis))
    ax1[i].plot(h_xaxis, hist/hist.sum(), 'o', ms=8, mfc='none', color=colors[i*2+1], markeredgewidth=2)
    ax1[i].set_ylabel('PDF', fontsize=16)
    ax1[i].set_xlabel('# of ' + Chunks_Name[i], fontsize=16)
    ax1[i].set_yscale('log')
    ax1[i].set_xscale('log')
    ax1[i].tick_params(axis='y', labelsize=12)
    ax1[i].tick_params(axis='x', labelsize=12)
    ax1[i].spines['top'].set_visible(False)
    ax1[i].spines['right'].set_visible(False)
plt.subplots_adjust(wspace=0.24, hspace=0.4)
```
