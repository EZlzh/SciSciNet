```python
sample_n = 50
import numpy as np
def CalculateCI(x, method):    
    data = x
    m = int(len(data)/10) # take 10% data
    ms = []
    for j in range(sample_n): # number of samples   
        if method=='mean':
                ms.append(np.mean(np.random.choice(data,m)))
        if method=='median':
                ms.append(np.median(np.random.choice(data,m)))
    a, b=[np.percentile(ms,2.5), np.percentile(ms,97.5)]
    return a, b
```

```python
%%time
# With bootstrapped CI
# It's not median citation, it's mean actually lol.
DD_Result = []
Dis_Result = []
Cit_Result = []
Da = []
Db = []
Ca = []
Cb = []
from tqdm import tqdm
df_Disrupt_Develop_GB = df_Disruption_Sel.groupby(['Team_Size'])
for name, df_group in df_Disrupt_Develop_GB:
#     print(name)
#     break
    if name > 10:
        continue
    Avg_Disrupt = df_group['Percentile'].median()
    Avg_Citation = df_group['Citation_Count'].mean()
    
    DD_Result.append([name, Avg_Disrupt, Avg_Citation])
    Dis_Result.append(Avg_Disrupt)
    Cit_Result.append(Avg_Citation)
    da, db = CalculateCI(df_group['Percentile'].values, 'median')
    ca, cb = CalculateCI(df_group['Citation_Count'].values, 'mean')
    print(name)
    print(Avg_Disrupt, Avg_Citation)
    print(da, db, ca, cb)
    Da.append(da)
    Db.append(db)
    Ca.append(ca)
    Cb.append(cb)
#     break
```
