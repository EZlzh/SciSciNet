```python
%%time
df_Ref_New = pd.read_csv('./Data-Metrics/MID_MAG_Journal_References_Year.csv', sep='\t')
df_Ref_New
```

```python
%%time
df_Ref_New['abs_year'] = df_Ref_New['Citing_Year'] - df_Ref_New['Cited_Year']
df_Ref_New
```

```python
%%time
df_Ref_Sel = df_Ref_New.loc[df_Ref_New['abs_year'] >=0, :]
df_Ref_Sel
```

```python
%%time
df_Ref_Sel = df_Ref_Sel.copy()
df_Ref_Sel
```

```python
import numpy as np
def sleeping_beauty(x):
#     print(x)
#     print("Begin Processing: ")
    
    time_list = x[:,0]
    cite_list = x[:,1]
    
#     print(time_list)
#     print(cite_list)
    
    max_time = int(max(time_list))
#     print(max_time)
    
    C_list = np.zeros((max_time+1,), dtype=int)
#     print(C_list)
    
    for i in range(len(time_list)):
        C_list[int(time_list[i])] = int(cite_list[i])
#     print(C_list)
    
    B_list = C_list.copy()
    B_list[C_list == 0] = 1
#     print(B_list)
#     print(C_list)
    
    t_m = np.argmax(C_list)
    
    if t_m == 0:
        return 0, 0
    
    c_m = C_list[t_m]
    c_0 = C_list[0]
#     print(c_m,t_m)
#     print(len(C_list), C_list.shape[0])

#     print(((c_m - c_0) / t_m * np.arange(t_m+1) + c_0 - C_list[:t_m+1]) / B_list[:t_m+1])
    B = np.sum(((c_m - c_0) / t_m * np.arange(t_m+1) + c_0 - C_list[:t_m+1]) / B_list[:t_m+1])
#     print(B)
    
    d_t = np.abs((c_m - c_0) * np.arange(t_m+1) + (c_0 - C_list[:t_m+1]) * t_m) / np.sqrt((c_m - c_0)**2 + t_m**2)
#     print(d_t)
    
    T = np.argmax(d_t)
    
    return B,T
#     return pd.Series([B, T],index=['BC','AT']).to_frame().transpose()
    
```

```python
%%time
df_Ref_Sel_Sample = df_Ref_Sel.loc[df_Ref_Sel['Cited'] <=500, :]
df_Ref_Sel_Sample.head(5)
```

```python
import numpy as np
from tqdm import tqdm
result = []
MAG_CC_Year_GBP = df_Ref_Sel_Sample.groupby(['Cited'], as_index=False, sort=False)
cnt = 0
for name, df_group in tqdm(MAG_CC_Year_GBP):
    B, T = sleeping_beauty(df_group.groupby('abs_year').size().reset_index(name='count').values)
    result.append([name, B, T])
    cnt += 1
print(result[:5])
```
