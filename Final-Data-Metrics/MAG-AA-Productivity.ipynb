```python
import pandas as pd
```

```python
%%time
SciNet_PaperAuthorAffiliations = pd.read_csv(
    '../Final-SciNet-Tables/SciNet_PaperAuthorAffiliations.tsv', sep='\t')
SciNet_PaperAuthorAffiliations
```

```python
%%time
SciNet_PaperAuthors_0 = SciNet_PaperAuthorAffiliations.drop(['AffiliationID','AuthorSequenceNumber'], axis=1).copy()
SciNet_PaperAuthors_0
```

```python
%%time
SciNet_PaperAuthors = SciNet_PaperAuthors_0.drop_duplicates()
SciNet_PaperAuthors
```

```python
Author_Sample_ID = [199142497, 2582258949]
df_A_Sample = SciNet_PaperAuthors.loc[SciNet_PaperAuthors['AuthorID'].isin(Author_Sample_ID),:].copy()
print(len(df_A_Sample))
df_A_Sample[:10]
```

```python
df_A_Sample_cnt = df_A_Sample.groupby(['AuthorID'])['PaperID'].agg(['count']).reset_index()
df_A_Sample_cnt
```

```python
%%time
df_Author_Prod = SciNet_PaperAuthors.groupby(['AuthorID'])['PaperID'].agg(['count']).reset_index()
df_Author_Prod
```

```python
%%time
df_Author_Prod.rename(columns={'count':'Productivity'}, inplace=True)
df_Author_Prod
```

```python
%%time
df_Author_Prod.to_csv('./Data-Count/SciNet-Prod.tsv', sep='\t', index=False)
```

```python
%%time
SciNet_PaperAuthorAffiliations_Sel = SciNet_PaperAuthorAffiliations.loc[
    pd.notnull(SciNet_PaperAuthorAffiliations['AffiliationID']), :]
SciNet_PaperAuthorAffiliations_Sel
```

```python
%%time
SciNet_PaperAffiliations_0 = SciNet_PaperAuthorAffiliations_Sel.drop(
    ['AuthorID','AuthorSequenceNumber'], axis=1).copy()
SciNet_PaperAffiliations_0
```

```python
%%time
SciNet_PaperAffiliations = SciNet_PaperAffiliations_0.drop_duplicates()
SciNet_PaperAffiliations
```

```python
%%time
df_Affiliation_Prod = SciNet_PaperAffiliations.groupby(['AffiliationID'])['PaperID'].agg(['count']).reset_index()
df_Affiliation_Prod
```

```python
df_Affiliation_Prod.rename(columns={'count':'Productivity'}, inplace=True)
df_Affiliation_Prod
```

```python
df_Affiliation_Prod.to_csv('./Data-Count/SciNet-Prod-Institute.tsv', sep='\t', index=False)
```
