```python
import pandas as pd
```

```python
%%time
SciNet_PaperAuthorAffiliations = pd.read_csv(
    '../Final-SciNet-Tables/SciNet_PaperAuthorAffiliations.tsv', sep='\t')
SciNet_PaperAuthorAffiliations
```

```python
%%time
SciNet_PaperAuthors_0 = SciNet_PaperAuthorAffiliations.drop(['AffiliationID','AuthorSequenceNumber'], axis=1).copy()
SciNet_PaperAuthors_0
```

```python
%%time
SciNet_PaperAuthors = SciNet_PaperAuthors_0.drop_duplicates()
SciNet_PaperAuthors
```

```python
%%time
SciNet_Papers = pd.read_csv('../Final-SciNet-Tables/SciNet_Papers-Temp0.tsv', sep='\t')
SciNet_Papers
```

```python
%%time
SciNet_PAYCC10 = SciNet_PaperAuthors.merge(
    SciNet_Papers[['PaperID','Year','Citation_Count','C10']], how='left', on='PaperID')
SciNet_PAYCC10
```

```python
import numpy as np
from tqdm import tqdm
tqdm.pandas()
```

```python
%%time
SciNet_PAYCC10_Sel = SciNet_PAYCC10.loc[pd.notnull(SciNet_PAYCC10['C10']), :]
SciNet_PAYCC10_Sel
```

```python
len(SciNet_PAYCC10_Sel['AuthorID'].unique())
```

```python
Author_Sample_ID = [1995014452, 2142263845]
df_A_Sample = SciNet_PAYCC10_Sel.loc[SciNet_PAYCC10_Sel['AuthorID'].isin(Author_Sample_ID),:].copy()
print(len(df_A_Sample))
df_A_Sample[:10]
```

```python
# Try a subset
Res_C10 = []
df_A_Sample_GB = df_A_Sample.groupby(
    ['AuthorID'], sort=False)
for name, df_G in tqdm(df_A_Sample_GB):
    C10_List = df_G['Citation_Count'].values
    C10_List_Norm = C10_List + 1
    cur_Avg_C10 = np.mean(C10_List)
    cur_Avg_LogC10 = np.mean(np.log(C10_List_Norm))
    Res_C10.append([name, cur_Avg_C10, cur_Avg_LogC10])
print(len(Res_C10))
Res_C10[:5]
```

```python
# Try a subset
Res_C10 = []
SciNet_PAYCC10_Sel_GB = SciNet_PAYCC10_Sel.groupby(
    ['AuthorID'], sort=False)
for name, df_G in tqdm(SciNet_PAYCC10_Sel_GB):
    C10_List = df_G['Citation_Count'].values
    C10_List_Norm = C10_List + 1
    cur_Avg_C10 = np.mean(C10_List)
    cur_Avg_LogC10 = np.mean(np.log(C10_List_Norm))
    Res_C10.append([name, cur_Avg_C10, cur_Avg_LogC10])
print(len(Res_C10))
Res_C10[:5]
```

```python
df_AvgC10 = pd.DataFrame(Res_C10)
df_AvgC10.rename(columns={0:'AuthorID',1:'Average_C10',2:'Average_LogC10'}, inplace=True)
df_AvgC10
```
