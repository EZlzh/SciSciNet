# H-index

```python
import numpy as np
from tqdm import tqdm
tqdm.pandas()
```

```python
def hindex(cit):
    # print(len(cit))
    # print(np.arange(cit.shape[0]) + 1)
    H_minus_R = np.sort(cit)[::-1] - np.arange(cit.shape[0])
    return (H_minus_R > 0).sum()
```

```python
Author_Sample_ID = [199142497, 2582258949]
df_A_Sample = SciNet_PAYCC10.loc[SciNet_PAYCC10['AuthorID'].isin(Author_Sample_ID),:].copy()
print(len(df_A_Sample))
df_A_Sample[:10]
```

```python
# Try a subset
df_A_Sample_out = df_A_Sample.groupby(
    ['AuthorID'], sort=False)['Citation_Count'].progress_apply(hindex).reset_index(name='H-index')
df_A_Sample_out
```

```python
# Try a subset
Res_H = []
df_A_Sample_GB = df_A_Sample.groupby(
    ['AuthorID'], sort=False)
for name, df_G in tqdm(df_A_Sample_GB):
    H_minus_P_List = np.sort(df_G['Citation_Count'].values)[::-1] - np.arange(len(df_G))
    cur_H = (H_minus_P_List > 0).sum()
    Res_H.append([name, cur_H])
print(len(Res_H))
Res_H[:5]
```

```python
%%time
Res_H = []
SciNet_PAYCC10_GB = SciNet_PAYCC10.groupby(
    ['AuthorID'], sort=False)
for name, df_G in tqdm(SciNet_PAYCC10_GB):
    H_minus_P_List = np.sort(df_G['Citation_Count'].values)[::-1] - np.arange(len(df_G))
    cur_H = (H_minus_P_List > 0).sum()
    Res_H.append([name, cur_H])
print(len(Res_H))
Res_H[:5]
```

```python
df_Hindex = pd.DataFrame(Res_H)
df_Hindex.rename(columns={0:'AuthorID',1:'H-Index'}, inplace=True)
df_Hindex
```
