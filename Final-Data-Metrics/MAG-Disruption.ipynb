```python
%%time
# Calculation of Disruption!
from tqdm import tqdm
D = {}

# cnt = 0

# Ref: Dict_Citing_to_Cited
# Cit: Dict_Cited_to_Citing

for paper in tqdm(Dict_Cited_to_Citing):
    # For each paper in Cited list, indicating at least one citation.
    if paper in Dict_Citing_to_Cited:
        # If this paper is also in Citing list, indicating at least one reference.
        S = set()
        for ref_of_paper in Dict_Citing_to_Cited[paper]:
            # For every ref_paper in paper's reference list.
            if ref_of_paper in Dict_Cited_to_Citing:
                # Add nk and nj into set S.  
                S.update(set(Dict_Cited_to_Citing[ref_of_paper]))
        
        n_j = len(set(Dict_Cited_to_Citing[paper]) & S)
        # nj by definition.
        
        n_i = len(Dict_Cited_to_Citing[paper]) - n_j
        # ni by definition.
        
        n_k = len(S) - n_j
        # nk by definition.
        
        D[paper] = (n_i - n_j) / (n_i + n_j + n_k)
        
        # Test:
        # print(n_i, n_j, n_k)
        # print(D[paper])
        
    else:
        # No reference, 
        # nj = nk = 0,
        # D[paper] = ni / ni = 1.
        
        D[paper] = 1.0
        
#     cnt += 1
#     if cnt >= 160000:
#         break
# print(cnt, len(D))
```
