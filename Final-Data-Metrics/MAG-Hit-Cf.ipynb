```python
df_MAG_PF_Year_Lv0_GBF = df_MAG_PF_Year_Lv0.groupby(['FieldID'],as_index=False)
```

```python
df_MAG_PF_Year_Lv0
```

```python
%%time
from tqdm import tqdm
import copy
Chunks = []
for field, df_group in tqdm(df_MAG_PF_Year_Lv0_GBF):
    # print(field)
    df_MAG_PF_Year_Lv0_GBFY = df_group.groupby(['Year'],as_index=False)
    # for year, df_G in tqdm(df_MAG_PF_Year_Lv0_GBFY):
    for year, df_G in df_MAG_PF_Year_Lv0_GBFY:
        # print(year)
        df_G = df_G.sort_values(by=['C_all'],ascending=False)
        len_df_G = len(df_G)
        
        pct = 0.01
        df_G['Lv0_1pct_Hit'] = 0

        pos = int(pct*len_df_G)
        if pos > 0:
            pos -= 1
        bound_c = df_G.iloc[pos, 3] # C_all is in the 3rd column.
        df_G.loc[df_G['C_all'] >= bound_c, 'Lv0_1pct_Hit'] = 1
        
        pct = 0.05
        df_G['Lv0_5pct_Hit'] = 0
        
        pos = int(pct*len_df_G)
        if pos > 0:
            pos -= 1
        bound_c = df_G.iloc[pos, 3] # C_all is in the 3rd column.
        df_G.loc[df_G['C_all'] >= bound_c, 'Lv0_5pct_Hit'] = 1
        
        
        pct = 0.10
        df_G['Lv0_10pct_Hit'] = 0
        
        pos = int(pct*len_df_G)
        if pos > 0:
            pos -= 1
        bound_c = df_G.iloc[pos, 3] # C_all is in the 3rd column.
        df_G.loc[df_G['C_all'] >= bound_c, 'Lv0_10pct_Hit'] = 1
        
        
        # Chunks.append(copy.deepcopy(df_G))
        Chunks.append(df_G)
        # break
    # break
# df_G
```

```python
%%time
df_MAG_Lv0_Hit = pd.concat(Chunks, ignore_index=True)
df_MAG_Lv0_Hit
```
