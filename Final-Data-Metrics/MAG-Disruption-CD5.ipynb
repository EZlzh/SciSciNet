```python
def Check_paper(paper, cur_year):
    if paper not in Dict_Paper_to_Year:
        return False
    tmp_year = Dict_Paper_to_Year[paper]
    if tmp_year >= cur_year and tmp_year - cur_year <= 5:
        return True
    return False
```

```python
%%time
# Calculation of Disruption!
from tqdm import tqdm
D = {}

# cnt = 0

# Ref: Dict_Citing_to_Cited
# Cit: Dict_Cited_to_Citing

for paper in tqdm(Dict_Cited_to_Citing):
    # Traverse each paper in Cited list, which indicates that it is at least with one citation.
    
    if paper not in Dict_Paper_to_Year:
        continue
    cur_year = Dict_Paper_to_Year[paper]
    
    if paper in Dict_Citing_to_Cited:
        # If this paper is also in Citing list, it is at least with one reference.
        S = set()
        for ref_of_paper in Dict_Citing_to_Cited[paper]:
            # For every ref_paper in paper's reference list.
            if ref_of_paper in Dict_Cited_to_Citing:
                # Add nk and nj into set S.  
                S.update(set(Dict_Cited_to_Citing[ref_of_paper]))
        
        new_S = set()
        for each_paper in S:
            if Check_paper(each_paper, cur_year) == True:
                new_S.add(each_paper)
        
        cur_Set = set()
        for each_paper in Dict_Cited_to_Citing[paper]:
            if Check_paper(each_paper, cur_year) == True:
                cur_Set.add(each_paper)
        
        
        n_j = len(cur_Set & new_S)
        # nj by definition.
        
        n_i = len(cur_Set) - n_j
        # ni by definition.
        
        n_k = len(new_S) - n_j
        # nk by definition.
        
        D[paper] = (n_i - n_j) / (n_i + n_j + n_k)
        
        # Test:
        # print(n_i, n_j, n_k)
        # print(D[paper])
        
    else:
        # No reference, 
        # nj = nk = 0,
        # D[paper] = ni / ni = 1.
        
        continue
        # D[paper] = 1.0
        
#     cnt += 1
#     if cnt >= 160000:
#         break
# print(cnt, len(D))
```
