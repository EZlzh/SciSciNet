```python
date_str = "2021-12-06"
```

```python
%%time
import csv
import pandas as pd

df_MAG = pd.read_csv('../../MAG_'+date_str+'/mag/Papers.txt', sep='\t', header=None, quoting=csv.QUOTE_NONE, 
                      usecols=[0,2,3,7,8,11,12,13,18,19,22,24])
# https://docs.microsoft.com/en-us/academic-services/graph/reference-data-schema#papers
# Read MAG data.
# 0:MAG_PaperId; 2:DOI; 3:DocType; 7:Year; 8:Date; 11:JournalId; 12:ConferenceSeriesId; 13:ConferenceInstanceId;
# 18:ReferenceCount; 19:CitationCount; 20:EstimatedCitation; 22:FamilyId; 24:DocSubTypes.

df_MAG
```

```python
%%time
SciNet_Papers = pd.read_csv('../Final-SciNet-Tables/SciNet_Papers-Temp0.tsv', sep='\t')
SciNet_Papers
```

```python
%%time
df_MAG_Sel = df_MAG.loc[df_MAG[0].isin(SciNet_Papers['PaperID']), :]
df_MAG_Sel
```

```python
import scipy.stats
```

```python
%%time
test_A_0 = df_MAG_Sel[18].values
test_A_1 = df_MAG_Sel[19].values
test_A_0
```

```python
%%time
test_A_P = scipy.stats.pearsonr(test_A_0, test_A_1)
test_A_P
```

```python
%%time
SciNet_Papers_Merge = SciNet_Papers.merge(df_MAG_Sel[[0,18,19]], how='left',left_on='PaperID',right_on=0)
SciNet_Papers_Merge
```

```python
%%time
SciNet_Papers_Merge = SciNet_Papers_Merge.drop(columns=[0])
SciNet_Papers_Merge
```

```python
%%time
SciNet_C = SciNet_Papers_Merge['Citation_Count'].values
MAGNet_C = SciNet_Papers_Merge[19].values
```

```python
%%time
C_Pearson = scipy.stats.pearsonr(SciNet_C, MAGNet_C)
C_Pearson
```

```python
%%time
C_Spearman = scipy.stats.spearmanr(SciNet_C, MAGNet_C)
C_Spearman
```

```python
%%time
SciNet_R = SciNet_Papers_Merge['Reference_Count'].values
MAGNet_R = SciNet_Papers_Merge[18].values
```

```python
%%time
R_Pearson = scipy.stats.pearsonr(SciNet_R, MAGNet_R)
R_Pearson
```

```python
%%time
R_Spearman = scipy.stats.spearmanr(SciNet_R, MAGNet_R)
R_Spearman
```
